```python
#establish path to all files in folder ending in ".csv"
path = "/data/OA3802g1/census/*.csv"
```

```python
#create empty df 
master_df = pd.DataFrame()
#loop over all files 
for fname in glob.glob(path):
    #make each file a new df
    df2 = pd.read_csv(fname)
    #change all the column names to their lowercase version
    df2.columns = map(str.lower, df2.columns)
    #add month worth of census data to master_df
    master_df = pd.concat([master_df, df2])
```

```python
#check work
master_df
```

```python
ad_df = pd.DataFrame()
for fname in glob.glob(path):
    df2 = pd.read_csv(fname)
    df2.columns = map(str.lower, df2.columns)
    #group df by if they are active duty or not and put series into df
    grouped_df = df2.groupby(['peafnow'], dropna=False).size().to_frame("Count")
    # calculate num of active duty members in census, total in census, and percentage based on those numbers
    data = [[grouped_df["Count"][1], df2.shape[0], grouped_df["Count"][1] / df2.shape[0]]]
    #make df3 these 3 new values
    df3 = pd.DataFrame(data, columns=['Active Duty', 'Num in Survey', 'Percentage Active Duty'])
    #add all month data to the ad_df
    ad_df = pd.concat([ad_df, df3])
```

```python
ad_df.reset_index(drop = True)
```

```python
#get mean of % column
fraction = ad_df['Percentage Active Duty'].mean()
#change to %
percent = fraction * 100
#print
print(f"Over the 34 census datasets, {percent:.2f}% of the sample identified as being active duty.")
```

```python
sex_df = pd.DataFrame()
for fname in glob.glob(path):
    df2 = pd.read_csv(fname)
    df2.columns = map(str.lower, df2.columns)
    #only look at active duty rows of monthly census
    filtered_df = df2.query('peafnow == 1')
    #group by sex
    grouped_df = filtered_df.groupby(['pesex'], dropna=False).size().to_frame("Count")
    data = [[grouped_df["Count"][1],
             grouped_df["Count"][2],
             filtered_df.shape[0], 
             (grouped_df["Count"][1] / filtered_df.shape[0])*100, (grouped_df["Count"][2] / filtered_df.shape[0])*100]]
    #make single row df number of active duty males, females, total active duty, and the percentages
    df3 = pd.DataFrame(data, columns=['AD Males', 'AD Females','Num AD', '%AD Male', '%AD Female'])
    #round calcs to two decimals
    df3 = df3.round(2)
    #add to sex_df
    sex_df = pd.concat([sex_df, df3])
```

```python
sex_df.reset_index()
```

```python
percentm = sex_df['%AD Male'].mean()
percentf = sex_df['%AD Female'].mean()
print(f"Over the 34 census datasets, {percentm:.2f}% of Active Duty sample was male and {percentf:.2f}% female.")
```

```python
race_df = pd.DataFrame()
for fname in glob.glob(path):
    print(fname)
    df2 = pd.read_csv(fname)
    df2.columns = map(str.lower, df2.columns)
    #only look at active duty members
    filtered_df = df2.query('peafnow == 1')
    #only keep the race column of ad members
    filtered_df2 = filtered_df[['ptdtrace']]
    #get rid of pesky warning
    pd.options.mode.chained_assignment = None
    #add 26 dummy rows to account for missing race observations in each monthy census
    for i in range(26):
        filtered_df2.loc[len(filtered_df2.index)+i] = [i+1]
    #group by race and get count in a df
    grouped_df = filtered_df2.groupby(['ptdtrace'], dropna=False).size().to_frame("Count")
    #subtract 1 off each value of race to remove dummy row counts... now the grouped_df has 0 for race categories not observed
    grouped_df.loc[:, "Count"] = grouped_df["Count"].apply(lambda x: x - 1)
    data = [[(grouped_df["Count"][1]/filtered_df.shape[0])*100,
         (grouped_df["Count"][2]/filtered_df.shape[0])*100,
         (grouped_df["Count"][3]/filtered_df.shape[0])*100,
         (grouped_df["Count"][4]/filtered_df.shape[0])*100,
         (grouped_df["Count"][5]/filtered_df.shape[0])*100,
         ((filtered_df.shape[0] - (grouped_df["Count"][1]+grouped_df["Count"][2]+grouped_df["Count"][3]+grouped_df["Count"][4]+grouped_df["Count"][5])) / filtered_df.shape[0])*100,
         filtered_df.shape[0]]]
    df3 = pd.DataFrame(data, columns=['White', 'Black','American Indian', 'Asian', 'Islander','Other', 'Total'])
    df3 = df3.round(1)
    race_df = pd.concat([race_df, df3])
```

```python
percentw = race_df['White'].mean()
percentb = race_df['Black'].mean()
percentai = race_df['American Indian'].mean()
percenta = race_df['Asian'].mean()
percentis = race_df['Islander'].mean()
percento = race_df['Other'].mean()
print(f"Over the 34 census datasets, {percentw:.2f}% of Active Duty sample was white.")
print(f"Over the 34 census datasets, {percentb:.2f}% of Active Duty sample was black.")
print(f"Over the 34 census datasets, {percentai:.2f}% of Active Duty sample was American Indian.")
print(f"Over the 34 census datasets, {percenta:.2f}% of Active Duty sample was Asian.")
print(f"Over the 34 census datasets, {percentis:.2f}% of Active Duty sample was Pacific Islander/Hawaiian.")
print(f"Over the 34 census datasets, {percento:.2f}% of Active Duty sample was Other.")
```

```python
hispanic_df = pd.DataFrame()
for fname in glob.glob(path):
    df2 = pd.read_csv(fname)
    df2.columns = map(str.lower, df2.columns)
    filtered_df = df2.query('peafnow == 1')
    filtered_df2 = filtered_df[['pehspnon']]
    grouped_df = filtered_df2.groupby(['pehspnon'], dropna=False).size().to_frame("Count")
    data = [[(grouped_df["Count"][1]/filtered_df2.shape[0])*100,
         (grouped_df["Count"][2]/filtered_df2.shape[0])*100,
         filtered_df.shape[0]]]
    df3 = pd.DataFrame(data, columns=['Hispanic', 'Non Hispanic','Total'])
    df3 = df3.round(1)
    hispanic_df = pd.concat([hispanic_df, df3])
```

```python
age_df = pd.DataFrame()
for fname in glob.glob(path):
    df2 = pd.read_csv(fname)
    df2.columns = map(str.lower, df2.columns)
    filtered_df = df2.query('peafnow == 1')
    filtered_df2 = filtered_df[['prtage']]
    filtered_df2 = filtered_df2.reset_index(drop=True)
    #use cut command to cut ages into groups and put "bin" as a new columns containing the age of the service member
    filtered_df2['bins'] = pd.cut(x=filtered_df2['prtage'], bins=[1, 25, 30, 35, 40, 100])
    #get rid of pesky warning
    warnings.simplefilter(action='ignore', category=FutureWarning)
    #continue as before
    grouped_df = filtered_df2.groupby(['bins']).size().to_frame("Count")
    count_list = grouped_df['Count'].tolist()
    data = [[(count_list[0]/filtered_df2.shape[0])*100,
         (count_list[1]/filtered_df2.shape[0])*100,
         (count_list[2]/filtered_df2.shape[0])*100,
         (count_list[3]/filtered_df2.shape[0])*100,
         (count_list[4]/filtered_df2.shape[0])*100,
         filtered_df2.shape[0]]]
    df3 = pd.DataFrame(data, columns=['<=25', '26-30','31-35','36-40','>=41','Total'])
    df3 = df3.round(1)
    age_df = pd.concat([age_df, df3])
```

```python
percent1 = age_df['<=25'].mean()
percent2 = age_df['26-30'].mean()
percent3 = age_df['31-35'].mean()
percent4 = age_df['36-40'].mean()
percent5 = age_df['>=41'].mean()
print(f"Over the 34 census datasets, {percent1:.2f}% of Active Duty sample was aged <=25.")
print(f"Over the 34 census datasets, {percent2:.2f}% of Active Duty sample was aged 26-30.")
print(f"Over the 34 census datasets, {percent3:.2f}% of Active Duty sample was aged 31-35.")
print(f"Over the 34 census datasets, {percent4:.2f}% of Active Duty sample was aged 36-40.")
print(f"Over the 34 census datasets, {percent5:.2f}% of Active Duty sample was aged >=41.")
```
